### First Question

```python
import time
start_time = time.time()
# First Question
final_df = tr_data.drop_duplicates(subset=['tx_hash']) # blk no is dropped duplicates dataframe
dff = final_df.groupby(["block_number","block_time"]).gas.sum().reset_index()
dff.sort_values(by = "gas", kind='heapsort')
dff.rename(columns={'gas': 'block_gas'}, inplace=True) # dff = first q
# dff.to_csv('/Users/srikaramara/Desktop/first_q.csv')
print("--- %s seconds ---" % (time.time() - start_time))
```

### Third Question 

```python
# Vectorization using numpy
final_df = final_df.reset_index()
total_gas_np = final_df.loc[:,'total_gas']
gas_np = final_df.loc[:, 'gas']

total_gas = total_gas_np.values
gas = gas_np.values

new_col = np.subtract(total_gas,gas)

final_df['transaction_fee'] = new_col.tolist()
thir_q = final_df.sort_values(by = "transaction_fee", kind='heapsort')
thir_q[['tx_hash','block_number','block_time','transaction_fee']]
```
